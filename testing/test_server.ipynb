```python
from civirank import analyzers, parsers, rankers

import json
import numpy as np

import time
import pickle

import warnings
with warnings.catch_warnings():
    warnings.simplefilter(action='ignore', category=FutureWarning)

import requests
import json
```

```python
def parse_constant(val):
    '''
    Parsing functions to get rid of NaN values that throw the requests lib
    See documentation of json.load() https://docs.python.org/3/library/json.html
    '''
    if val in ["NaN", "Infinity", "-Infinity"]:
        return ""
    else:
        return val


# load sample data
fname = "twitter_test.json"
with open(fname, "r") as fin:
    sample_data_twitter = json.load(fin, parse_constant=parse_constant)
fname = "reddit_test.json"
with open(fname, "r") as fin:
    sample_data_reddit = json.load(fin, parse_constant=parse_constant)
fname = "facebook_test.json"
with open(fname, "r") as fin:
    sample_data_facebook = json.load(fin, parse_constant=parse_constant)
```

```python
# Define the URL for the API endpoint
url = "http://127.0.0.1:8000/rank"

# Define the payload


# Set the headers
headers = {
    "Content-Type": "application/json"
}


# Send the request and capture the response
response = requests.post(url, json = sample_data_facebook)

# Check the response status code and content
if response.status_code == 422:
    print("Error:")
else:
    print("Success:")
```
